```python
curr,conn = create_database()
```

```python
customers_table_create = ("""
        CREATE TABLE IF NOT EXISTS customers (
            customer_id INTEGER PRIMARY KEY,
            customer_name VARCHAR(255),
            gender VARCHAR(30),
            age INTEGER,
            home_address VARCHAR(500),
            zip_code VARCHAR(10),
            city VARCHAR(255),
            state VARCHAR(255)
        );
    """)
```

```python
curr.execute(customers_table_create)
```

```python
conn.commit()
```

```python
products_table_create = ("""
        CREATE TABLE IF NOT EXISTS products (
            product_id INTEGER PRIMARY KEY,
            product_type VARCHAR(255),
            product_name VARCHAR(255),
            size VARCHAR(10),
            colour VARCHAR(50),
            price INTEGER,
            quantity INTEGER        
        );
    """)
```

```python
curr.execute(products_table_create)
```

```python
conn.commit()
```

```python
orders_table_create=("""
        CREATE TABLE IF NOT EXISTS orders (
            order_id INTEGER PRIMARY KEY,
            customer_id INTEGER,
            payment INTEGER,
            order_date DATE,
            delivery_date DATE
        );
    """)
```

```python
curr.execute(orders_table_create)
conn.commit()
```

```python
sales_table_create = ("""
        CREATE TABLE IF NOT EXISTS sales (
            sales_id INTEGER PRIMARY KEY,
            order_id INTEGER,
            product_id INTEGER,
            price_per_unit INTEGER,
            quantity INTEGER,
            total_price INTEGER
        );
    """)
```
